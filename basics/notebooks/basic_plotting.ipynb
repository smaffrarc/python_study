```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 100, 0.1)
```

```python
y = np.sin(x)
```

```python
plt.figure(figsize=(10, 1))
plt.ylim(-2,2)
plt.plot(x, y, color='r', lw=1);
```

```python
def myfunc(x):
    temp = []
    for value in x:
        new_value = np.sin(value)
        if new_value < 0.0:
            new_value = 0.0
        temp.append(new_value)
    return np.array(temp)
        
```

```python
z = myfunc(x)
```

```python
plt.figure(figsize=(10, 1))
plt.ylim(-2,2)
plt.plot(x, z, color='blue', lw=1);
```
